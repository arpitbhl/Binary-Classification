```python
# importing useful libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.svm import SVC
# Importing Training set
data = pd.read_csv('training_set.csv')
dataset = pd.DataFrame(data)
dataset.columns
```

```python
dataset.info()
```

```python
dataset.drop(dataset.columns[[0]],axis=1,inplace=True)
dataset.info()
```

```python
dataset.describe()
```

```python
dataset = dataset.replace('?', np.nan)
dataset = dataset.apply(lambda x: x.fillna(x.median()),axis=0)

dataset.isnull().sum()
```

```python
# Corelation Matrix
matrix = dataset.corr()
t = matrix.iloc[57].to_numpy().tolist()
column = list(matrix)
print(t,column)
```

```python
drop = []
for i in range(len(t)):
    if(t[i]<0.01 and t[i]>-0.2):
        drop.append(column[i])
drop.append('Y')
print(drop)
```

```python
target = dataset['Y']
features = dataset.drop(drop, axis=1)
X_train, X_test, y_train, y_test = train_test_split(features,target, test_size = 0.2, random_state = 10)
model = LogisticRegression(solver='liblinear', random_state=0).fit(X_train, y_train)
prediction = model.predict(X_test)
```

```python
print("Confusion Matrix:\n",confusion_matrix(prediction,y_test))
print("Accuracy Score:\n",accuracy_score(y_test,prediction))
```

```python
# Now we try for the test dataset
data = pd.read_csv('test_set.csv')
dataset = pd.DataFrame(data)
dataset.columns
dataset.drop(dataset.columns[[0]],axis=1,inplace=True)
dataset.info()
```

```python
if 'Y' in drop:
    drop.remove('Y')
features = dataset.drop(drop, axis=1)
predict_test = model.predict(features).tolist()
print(predict_test)
df = pd.DataFrame()
df['Y'] = predict_test
df.to_csv('predict_test.csv')
```
